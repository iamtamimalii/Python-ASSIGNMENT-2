```python
# cLustring is unsupervise lerarning  
# cLustring is group( set of items) 
# we have 4 things -> 1)K 2)X 3)C 4)Formula
# k shows the number of groups to be formed 
# in case of Classes (X) we sepreate the students in male and female group (C1 & C2) then the value of k will be 2 
# formula to calculate the grouping 

# steps to follow:
# decice the value of K
# get data (X) & (C1, C2)
# -> SQRT((x-c)^2) 
# -> |x-c| 
# the answer we get from the formula decide in which group data should be placed
# now take average of all grouped data to get the new point 
# it run as many as the data became constant 
# now repeate the process to get final answer 
```

```python
X = [int(x) for x in input("\n Enter X values (with space): ").split()] 
    
C1, C2 = [int(x) for x in input("\n Enter random value for C1 and C2: ").split()]
```

```python
size=len(X)
Flag = False
iterations = 0

while Flag != True:
    grp_C1 = []
    grp_C2 = []
    grp_C1_min = []
    grp_C2_min = []
    iteration_C1 = C1 
    iteration_C2 = C2
    temp1 = temp2 = 0
    count1 = count2 = 0
    avg_C1 = avg_C2 = 0
    
    for i in range (size):
        temp1 = X[i] - C1
        temp2 = X[i] - C2
        
        if temp1 < 0:
            temp1 *= (-1)
        
        if temp2 < 0:
            temp2 *= (-1)
        
        temp1 = round(temp1,2)
        grp_C1.append(temp1)
        temp2 = round(temp2,2)
        grp_C2.append(temp2)
        
        if temp1 < temp2:
            avg_C1 += X[i]
            grp_C1_min.append(X[i])
            count1 += 1
        else:
            avg_C2 += X[i]
            grp_C2_min.append(X[i])
            count2 += 1
        
    if avg_C1 != 0:
        avg_C1 /= count1
        avg_C1 = round(avg_C1,2)
    
    if avg_C2 != 0:
        avg_C2 /= count2
        avg_C2 = round(avg_C2,2)

    if C1 == avg_C1 and C2 == avg_C2:
        Flag = True
    else:
        C1 = avg_C1
        C2 = avg_C2
    
    C1 = round(C1,2)
    C2 = round(C2,2)
    
    print("\n\n\n", iterations, "Iteration Where C1 =", iteration_C1,"And C2 =", iteration_C2)
    print("_______________________________")
    print("|   X   |    C1    |    C2    |")
    print("-------------------------------")
    
    iterations += 1
    
    for i in range (size):
        print("|", X[i], " "*(4-len(str(X[i]))), "|",
              grp_C1[i], " "*(7-len(str(grp_C1[i]))), "|",
              grp_C2[i], " "*(7-len(str(grp_C2[i]))), "|")
        
    print("_______________________________")
    print("Group1: ", grp_C1_min, "\nAverage of group1: ", avg_C1)
    print("\nGroup2: ", grp_C2_min, "\nAverage of group2: ", avg_C2)
```

```python
if C1 == avg_C1 and C2 == avg_C2:
    print("Both", iterations - 1, "iteration &", iterations, "iteration become same. ðŸ˜")
else:
    print("Both", iterations - 1, "iteration &", iterations, "iteration are not becoming same. ðŸ˜¢")
```
